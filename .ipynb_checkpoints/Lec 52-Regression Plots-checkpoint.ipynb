```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
#Loas seaborn example dataset
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,ci=75)
#can specify confidence interval
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#Can not fit a regression
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
#lmplot() also works on discrete variables, like percentage

#New column for tip percentage
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])

sns.lmplot('size','tip_pect',tips)
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
#Adds 
```

```python
#Can estimate tendency of bin
#Size of party in this case
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
#can use the hue facet to automatically define subsets along a column

sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
#Plot by day
```

```python
#Can do LOESS model fitting
sns.lmplot('total_bill','tip_pect',tips,lowess=True,
           line_kws={'color':'black'})
```

```python
# The lmplot() is actually using a lower-level function regplot()
sns.regplot('total_bill','tip_pect',tips)
```

```python
# Create figure with 2 subplots
fig, (axis1,axis2) = plt.subplots(1,2,sharey =True)

sns.regplot("total_bill","tip_pect",tips,ax=axis1)
sns.violinplot(x="size",y="tip_pect",data=tips.sort('size'),palette="Set3")
```
