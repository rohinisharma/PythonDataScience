```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(100)

sns.distplot(dataset,bins=25)
```

```python
sns.distplot(dataset,rug=True,hist=False)
```

```python
# TO control specific plots in distplot
#use [plot]_kws argument with dictionaries.

sns.distplot(dataset,bins=25,
            kde_kws ={'color':'indianred','label':'KDEPLOT'},
            hist_kws ={'color':'blue','label':'HISTOGRAM'})
```

```python
from pandas import Series

ser1 = Series(dataset,name='MYDATA')
```

```python
sns.distplot(ser1,bins=25)
```
