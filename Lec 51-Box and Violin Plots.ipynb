```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.boxplot(data=data1,vert=False,whis=np.inf)
```

```python
#Normal distribution data
data1 = stats.norm(0,5).rvs(100)

#2 gamma distributions, concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

#Boxplot data1 and data2

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2])
#Shows where data is most dense
```

```python
sns.violinplot(data=data2,bw=.01)
```

```python
sns.violinplot(data=data1,inner='stick')
```
