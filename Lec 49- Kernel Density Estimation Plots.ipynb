```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(25)
#Rug plot puts tick where a value occurs
sns.rugplot(dataset)
#Y-axis limit
```

```python
plt.hist(dataset,alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset);
#Set up x-axis
x_min = dataset.min() -2
x_max = dataset.max() +2
#100 equally spaced points between x_min and x_max
x_axis = np.linspace(x_min,x_max,100)
#Sets up bandwith using formula
bandwith = ((4*dataset.std()**5)/(3*len(dataset)))**.2

#Empty list 
kernel_list = []

for data_point in dataset:
    
    #Create a kernel for each data poing and append to list
    kernel = stats.norm(data_point,bandwith).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel/kernel.max()
    kernel = kernel *.4
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
#To get KDE sum these basic functions

#Sum of basis function
sum_of_kde = np.sum(kernel_list,axis=0)

#Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

#Add initial rugplot
sns.rugplot(dataset,c='indianred')

#Get rid of y ticks
plt.yticks([])

#Title
plt.suptitle('Sum of basis functions')
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset,color='black')

#Plot various bandwiths
for bw in np.arange(.5,2,.25):
    sns.kdeplot(dataset,bw=bw,lw=1.8,label=bw)
```

```python
#Can chooses different kernel types
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern)
```

```python
# We can also shade if desired
for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern,shade=True,alpha=0.5)
```

```python
sns.kdeplot(dataset,vertical=True)
```

```python
sns.kdeplot(dataset,cumulative=True)
```

```python
#Plotting multivariate density estimation

#Mean center of data
mean = [0,0]

cov=[[1,0],[0,100]]
#Create dataset
dataset2 = np.random.multivariate_normal(mean,cov,1000)

dframe = pd.DataFrame(dataset2,columns=['X','Y'])

sns.kdeplot(dframe)
```

```python
#Pass 2 vectors seperately
sns.kdeplot(dframe.X,dframe.Y,shade=True)
```

```python
sns.kdeplot(dframe,bw=1)
```

```python
sns.kdeplot(dframe,bw='silverman')
```

```python
sns.jointplot('X','Y',dframe,kind='kde')
```
